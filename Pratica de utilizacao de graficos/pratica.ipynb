```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tabela = pd.read_csv("dados.csv",sep=";")
display(tabela)
```

```python
sns.histplot(tabela.PIB, kde=True, bins=10).set(title='PIB')
```

```python
sns.histplot(tabela.VALOREMPENHO,color="green",bins=10).set(title='VALOR EMPENHO')
```

```python
sns.kdeplot(tabela.VALOREMPENHO,color="blue").set(title = "Densidade Valor Empenho")
```

```python
dados_selec = tabela[["PIB","VALOREMPENHO"]]
dados_selec.hist(bins=20, alpha=0.7, figsize=(15, 5))
```

```python
selec_aleatoria = tabela.sort_values('PIB').head(10)
selec_aleatoria.plot.bar(x='MUNICIPIO',y='PIB', color = 'gray')
```

```python
selec_aleatoria = tabela.sort_values("VALOREMPENHO").head(10)
selec_aleatoria.plot.bar(x="MUNICIPIO",y="VALOREMPENHO", color = "blue")
```

```python
tabela2 = pd.read_csv("trees.csv")
tabela2.head()
```

```python
plt.scatter(tabela2.Girth,tabela2.Volume,marker="*")
plt.title("Arvores Volume x Circunferencia")
plt.xlabel("Circunferencia")
plt.ylabel("Volume")
```

```python
sns.regplot(x=tabela2.Volume, y=tabela2.Girth, data= tabela2,x_jitter=0.4, fit_reg=False)
```

```python
plt.boxplot(tabela2.Volume, vert = False, notch= True , patch_artist= True)
```

```python
sns.boxenplot(data=tabela2, orient="h")
```

```python
plt.figure()
plt.subplot(3,2,1)
sns.histplot(tabela.PIB, kde=True, bins=10).set(title='PIB')
plt.subplot(3,2,2)
sns.kdeplot(tabela.VALOREMPENHO,color="blue").set(title = "Densidade Valor Empenho")
plt.subplot(3,2,3)
sns.regplot(x=tabela2.Volume, y=tabela2.Girth, data= tabela2,x_jitter=0.4, fit_reg=False)
plt.subplot(3,2,4)
sns.boxenplot(data=tabela2, orient="h")
plt.tight_layout()
```
